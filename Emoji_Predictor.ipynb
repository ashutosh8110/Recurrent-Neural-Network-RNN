```python
X_train=data[:,0]
Y_train=data[:,1]
X_test=test['Sentence'].values
Y_test=test['Emoji'].values
```

```python
from keras.utils import to_categorical
```

```python
Y_test=to_categorical(Y_test)
Y_train=to_categorical(Y_train)
```

## GET EMBEDDINGS

```python
glove_vec={
          }
with open('glove.txt',encoding='mbcs') as f:
    doc=f.readlines()
    
for i in range(len(doc)):
    word   = doc[i].split(" ")[0]
    vector = np.array(doc[i].split(" ")[1:],dtype='float64')
    
    glove_vec[word]=vector
```

```python
length={}
for i in range(X_train.shape[0]):
    d=len(X_train[i].split())
    if(d>10):
        print(X_train[i])
    if(d in length):
        length[d]+=1
        
    else:
        length[d]=1
        
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(list(length.keys()),list(length.values()))
```

```python
def get_Output_Embeddings(x):
    maxlen=10
    embedding_matrix=np.zeros((x.shape[0],maxlen,50))
    for i in range(x.shape[0]):
        words=x[i].split()
        for j in range(len(words)):
            embedding_matrix[i,j]=glove_vec[words[j].lower()]
            
    return embedding_matrix
```

```python
X_train[0]
```

```python
X_train= get_Output_Embeddings(X_train)
X_test = get_Output_Embeddings(X_test )
```

```python
X_train.shape,X_test.shape
```

```python
from keras.layers import *
from keras.models import Sequential
```

```python
model=Sequential()
model.add(LSTM(64,input_shape=(10,50),return_sequences=True))
model.add(Dropout(0.3))
model.add(LSTM(64))
model.add(Dense(5,activation='softmax'))
model.summary()
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['acc'])
```

```python
model.fit(X_train,Y_train,epochs=18,batch_size=32,shuffle=True,validation_split=0.1)
```

```python
Y_pred=model.predict_classes(X_test)
```

```python
model.evaluate(X_test,Y_test)
```
